#### Feature selection is a crucial step in the model preparation process to enhance the performance and efficiency of machine learning models by identifying and utilizing the most relevant features in a dataset. Three popular methods for feature selection include the Correlation Matrix method, Extra Trees Classifier method, and SelectKBest method.

<b>1- Correlation Matrix Method:</b>

* The Correlation Matrix method involves analyzing the correlation between each pair of features in the dataset. Features with high correlation may contain redundant information, and selecting one from such pairs can improve model interpretability and reduce overfitting.
* By examining the correlation matrix, features that have a strong correlation with the target variable while having low intercorrelation can be considered important for model building.

<b>2 - Extra Trees Classifier Method:<b/>

* The Extra Trees Classifier method employs an ensemble of decision trees to evaluate the importance of each feature in predicting the target variable.
* By assessing the information gain or decrease in impurity for each feature during the construction of decision trees, this method assigns importance scores to features. Features with higher importance scores are deemed more relevant for the model.

<b>3 - SelectKBest Method:<b/>

* The SelectKBest method is a univariate feature selection technique that evaluates the statistical significance of each feature independently with respect to the target variable.
* Common statistical tests such as chi-squared, ANOVA, or mutual information are applied to assign scores to features. The top 'k' features with the highest scores are then selected for model training.


```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset into a pandas DataFrame
data = pd.read_csv('global_heat_index.csv')

# Display the first few rows of the dataset to understand its structure
data.head()
```

```python
# Display information about the dataset
data.info()
```

## <b/> Cleaning Data

```python
# Filter data for records with an hour before 6:00 AM
clean1 = data[data['Hour'] < 6]

# Filter data for records with an hour after 6:00 PM
clean2 = data[data['Hour'] > 18]

# Display the cleaned data for the specified time range (12:00 AM to 5:59 AM)
print('Cleaning Data from 12:00 AM to 5:59 AM')
clean1.head()
```

```python
# Remove rows where the hour is before 6:00 AM
data = data.drop(clean1.index, axis=0)

# Remove rows where the hour is after 6:00 PM
data = data.drop(clean2.index, axis=0)
```

```python
# Check the of the dataset now
data.shape
```

## <b/> Input and Output Data

```python
# Extract the first 10 columns as feature variables (assuming your features are in the first 10 columns)
x = data.iloc[:, 0:10]

# Extract the last column as the target variable
y = data.iloc[:, -1]
```

```python
# Import the MinMaxScaler from scikit-learn
from sklearn.preprocessing import MinMaxScaler

# Create an instance of MinMaxScaler and use it to transform and scale the feature variables (x)
scaler = MinMaxScaler()
x = pd.DataFrame(scaler.fit_transform(x))
```

```python
# Check the scaled data
x.head()
```

# <b/><font color='green'>1. Correlation Matrix with Heatmap</font>

```python
# Calculate the correlation matrix for the dataset
corr_matrix = data.corr()
```

```python
# See the Correlation Matrix
corr_matrix
```

```python
# Plot the correlation matrix using heatmap
plt.figure(figsize=(12,10))
sns.heatmap(data=corr_matrix, annot=True)
```

# <b/><font color='green'>2. Extra Tree Classifier</font>

```python
# Import the ExtraTreesClassifier from scikit-learn
from sklearn.ensemble import ExtraTreesClassifier

# Create an instance of ExtraTreesClassifier
model = ExtraTreesClassifier()

# Fit the model to your data
model.fit(x, y)
```

```python
# Specify the column names for your features
col = ['Year', 'Month', 'Day', 'Hour', 'Dew Point', 'Temperature', 'Pressure',
       'Relative Humidity', 'Wind Direction', 'Wind Speed']

# Create a pandas Series with feature importances, using column names as indices
feature_imp = pd.Series(model.feature_importances_, index=col)

# Display the feature importances
feature_imp
```

```python
# Plot the top 10 features with the highest importance using a bar plot
feature_imp.nlargest(10).plot(kind='bar', edgecolor='black')

# Set plot labels and title
plt.xlabel('Features')
plt.ylabel('Importance Score')
plt.title('Top 10 Features - Extra Trees Classifier')

# Show the plot
plt.show()
```

# <b/><font color='green'>3. SelectKBest Method</font>

```python
# Import the necessary modules for feature selection
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
# Create an instance of SelectKBest with the chi-squared test as the scoring function and specify the number of features (k)
best_features = SelectKBest(score_func=chi2, k=10)

# Fit the SelectKBest instance to your data
best_features.fit(x, y)
```

```python
# Create a DataFrame for the chi-squared scores
scores = pd.DataFrame(best_features.scores_)

# Create a DataFrame for the column names
columns = pd.DataFrame(col)
```

```python
# Concatenate the DataFrames horizontally
feature_score = pd.concat([columns, scores], axis=1)

# Rename the columns
feature_score.columns = ['Features', 'Scores']

# Display the DataFrame with feature names and their corresponding chi-squared scores
feature_score
```

```python
# Plot the top 10 features with the highest chi-squared scores using a bar plot
feature_score.nlargest(10, 'Scores').plot(kind='bar', x='Features', y='Scores', edgecolor='black')

# Set plot labels and title
plt.xlabel('Features')
plt.ylabel('Chi-squared Scores')
plt.title('Top 10 Features - SelectKBest with Chi-squared Test')

# Show the plot
plt.show()
```
