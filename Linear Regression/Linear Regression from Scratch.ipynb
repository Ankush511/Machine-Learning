# <font color='green'>Linear Regression</font>

### Forward Propagation

f(x) = m*x + c

```python
def forward_propagation(train_input, parameters):
    m = parameters['m']
    c = parameters['c']

    predictions = np.multiply(m, train_input) + c

    return predictions
```

- This function forward_propagation performs forward propagation in a simple linear regression model.

### Cost Function

summation(1/2n * (y - f(x))^2)

```python
def cost_function(predictions, train_output):
    
    cost = np.mean((train_output - predictions) ** 2) * 0.5

    return cost
```

- This function cost_function calculates the cost (or loss) associated with the predictions made by the model compared to the actual output values. 

### Gradient Descent / Backward Propagation

df = summation(f(x) - y) / n

dm = df * x

dc = df * 1

```python
def backward_propagation(train_input, train_output, predictions):
    derivatives = dict()

    df = predictions - train_output
    dm = np.mean(np.multiply(df, train_input))
    dc = np.mean(df)

    derivatives['dm'] = dm
    derivatives['dc'] = dc

    return derivatives
```

- This function backward_propagation calculates the derivatives of the cost function with respect to the parameters of the model (slope and intercept) using the method of backward propagation. 

### Update Parameters

m = m - (learning_rate * dm)

c = c - (learning_rate * dc)

```python
def update_parameters(parameters, derivatives, learning_rate):
    
    parameters['m'] = parameters['m'] - (learning_rate * derivatives['dm'])
    parameters['c'] = parameters['c'] - (learning_rate * derivatives['dc'])

    return parameters
```

- This function update_parameters is responsible for updating the parameters of the linear regression model based on the calculated derivatives and a specified learning rate.

## Model Training

```python
def train(train_input, train_output, learning_rate, iters):
    # random parameters
    parameters = dict()
    parameters['m'] = np.random.uniform(0,1) * -1
    parameters['c'] = np.random.uniform(0,1) * -1

    plt.figure()

    # loss
    loss = list()

    # iteration
    for i in range(iters):
        # forward propagation
        predictions = forward_propagation(train_input, parameters)

        # cost
        cost = cost_function(predictions, train_output)

        loss.append(cost)
        print(f'Iteration : {i+1}, Loss : {cost}')

        # plot
        fig, ax = plt.subplots()
        
        ax.plot(train_input, train_output, '+', label='Original')
        ax.plot(train_input, predictions, '*', label='Training')
        
        legend = ax.legend()

        plt.plot(train_input, train_output, '+', label='Original')
        plt.plot(train_input, predictions, '*', label='Training')
        plt.show()

        # backward propagation
        derivatives = backward_propagation(train_input, train_output, predictions)

        # update parameters
        parameters = update_parameters(parameters, derivatives, learning_rate)

    return parameters, loss
    
```

- This train function performs training for a simple linear regression model using gradient descent. Overall, this function implements the training process for a linear regression model, visualizing the training progress and returning the trained parameters along with the loss values.

### Training Input & Output

```python
parameters, loss = train(train_input, train_output, 0.0001, 20)
```

- Here we invoke the train function to train a linear regression model using the provided training input and output data.
