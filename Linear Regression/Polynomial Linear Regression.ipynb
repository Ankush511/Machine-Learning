# <b/><font color='green'>Polynomial Linear Regression</font>

```python
# importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# loading data
data = pd.read_csv('Position_Salaries.csv')
data.head()
```

```python
# check rows & columns
data.shape
```

### <b/>Splitting the Data

```python
x = data.iloc[:,1:-1].values
y = data.iloc[:,2].values
```

```python
x
```

```python
y
```

## <b/><font color='green'>Linear Regression</font>

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(x,y)
```

### <b/>Visualizing the LR model

```python
plt.scatter(x,y, color='red')
plt.plot(x, lr.predict(x), color='blue')
plt.xlabel('Input')
plt.ylabel('Prediction')
plt.show()
```

## <b/><font color='green'>Polynomial Regression Features</font>

```python
from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree=4)
x_poly = poly_reg.fit_transform(x)
```

```python
x_poly
```

```python
poly_linear_reg = LinearRegression()

poly_linear_reg.fit(x_poly, y)
```

### <b/>Visualizing Polynomial LR Model

```python
plt.scatter(x, y, color='red')
plt.plot(x, poly_linear_reg.predict(x_poly), color='blue')
plt.title('Polynomial Regression with degree = 2')
plt.xlabel('Input')
plt.ylabel('Predictions')
plt.show()
```

```python
plt.scatter(x, y, color='red')
plt.plot(x, poly_linear_reg.predict(x_poly), color='blue')
plt.title('Polynomial Regression with degree = 3')
plt.xlabel('Input')
plt.ylabel('Predictions')
plt.show()
```

```python
plt.scatter(x, y, color='red')
plt.plot(x, poly_linear_reg.predict(x_poly), color='blue')
plt.title('Polynomial Regression with degree = 4')
plt.xlabel('Input')
plt.ylabel('Predictions')
plt.show()
```
