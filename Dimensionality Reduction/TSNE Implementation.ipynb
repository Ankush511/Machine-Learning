```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load dataset
data = pd.read_csv('bank_note_data.csv')
data.head()
```

```python
# unique elements in class column
data['Class'].unique()
```

```python
# data information
data.info()
```

```python
# target and features 
features = list(data.columns)[:-1]
target = list(data.columns)[-1]
```

## <b/> Scaling Data

```python
from sklearn.preprocessing import StandardScaler

data_norm = data.copy()
sc = StandardScaler()
data_norm[features] = sc.fit_transform(data[features])
```

```python
data_norm
```

# <b/><font color='green'>TSNE</font>

```python
from sklearn.manifold import TSNE

tsne = TSNE(learning_rate=500, n_components=2)

x_tsne = tsne.fit_transform(data_norm[features])
```

```python
x_tsne
```

```python
y = data[target]
```

## <b/>Visualizations

```python
plt.figure(figsize=(12,8))

plt.scatter(x_tsne[np.where(y==0) ,0], x_tsne[np.where(y==0), 1], c='green', marker='*')
plt.scatter(x_tsne[np.where(y==1) ,0], x_tsne[np.where(y==1), 1], c='red', marker='^')

plt.xlabel('Dimension_1')
plt.ylabel('Dimension_2')
plt.title('TSNE Plot')
plt.show()
```
