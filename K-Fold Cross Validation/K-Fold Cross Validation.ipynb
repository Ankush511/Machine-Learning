```python
# importing libraires
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.axes as ax

import warnings
warnings.filterwarnings('ignore')
```

```python
# load the data
data = pd.read_csv('boston_house_prices.csv')
data.head()
```

## <b/> Splitting the data

```python
x = data.drop(['MEDV'], axis=1)
y = data['MEDV']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

## <b/>Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(100)
rf = rf.fit(X_train, y_train)

predictions = rf.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_test, predictions)
```

# <b/><font color='green'>K-Fold Cross Validation</font>

```python
from sklearn.model_selection import KFold

kf = KFold(n_splits=5, shuffle=False)

for train_index, test_index in kf.split(x):
    print("Train - \n", train_index, '\nTest - \n', test_index)
    print("-----------------------")
```

```python
# again training the random forest model with k-fold data
rf1 = RandomForestRegressor()

n_est = [100,200,500]

for n_estimator in n_est:
    errors = []
    for train_index, test_index in kf.split(x):
        X_train, X_test = x.iloc[train_index], x.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]

        rf1 = RandomForestRegressor(n_estimator)
        rf1 = rf1.fit(X_train, y_train)
        pred = rf1.predict(X_test)

        mse = mean_squared_error(y_test, pred)
        errors.append(mse)

    print('Results for n_estimators : ', n_estimator)
    print(errors)
    print('Average Error : ', sum(errors)/len(errors))
```
