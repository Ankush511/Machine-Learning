```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

import warnings
warnings.filterwarnings('ignore')
```

```python
# making data points
X,y = make_blobs(n_samples=500, n_features=2, centers=5, random_state=3)
```

```python
X.shape
```

```python
# plotting the data points
fig = plt.figure(0)
plt.grid(True)
plt.scatter(X[:,0],X[:,1], c=y)
plt.show()
```

# <b/><font color='green'>K-Means Clustering</font>

### <b/> Step 1: Random Initialization of Centers

```python
# defining k
k=5

# defining cluster colour 
colors = ['green','yellow','blue','cyan','red']

clusters = {}

for i in range(k):
    center = 10*(2*np.random.random((X.shape[1],)) - 1)
    points = []
    cluster = {
        'center': center,
        'points': [],
        'color': colors[i]
    }

    clusters[i] = cluster
```

```python
clusters
```

```python
plt.scatter(X[:,0], X[:,1], c=y)

for i in clusters:
    center = clusters[i]['center']
    plt.scatter(center[0],center[1],marker='^', c='red')

plt.show()
```

### <b/> Step 2: Assigning each point to the nearest center

```python
def distance(v1,v2):
    return np.sqrt(np.sum((v1-v2)**2))
```

```python
# Implementing the E - step now
def assign_clusters():
    for i in range(X.shape[0]):
        dist = []

        curr_x = X[i]

        for j in range(k):
            dis = distance(curr_x, clusters[j]['center'])
            dist.append(dis)
        curr_cluster = np.argmin(dist)
        clusters[curr_cluster]['points'].append(curr_x)
```

```python
assign_clusters()
```

```python
clusters
```

```python
def plot_clusters():
    for i in clusters:
        pts = np.array(clusters[i]['points'])

        try:
            plt.scatter(pts[:,0], pts[:,1], c=clusters[i]['color'])
        except:
            pass

        center = clusters[i]['center']
        plt.scatter(center[0],center[1],c='black', marker='^')
```

```python
plot_clusters()
```

### <b/> Step 3: Update the center location by taking the mean

```python
# Implementing the M - step

def update_cluster():
    for i in range(k):
        pts = np.array(clusters[i]['points'])

        if pts.shape[0] > 0:
            new_center = pts.mean(axis=0)
            clusters[i]['center'] = new_center

            clusters[i]['points'] = []
```

```python
assign_clusters()
plot_clusters()
update_cluster()
```

# <b/> Optimizing K-Means Clustering using Sklearn Library

```python
from sklearn.cluster import KMeans
```

```python
# initializing K-means with 5 clusters
km = KMeans(n_clusters=5)
```

```python
# making data points
X,y = make_blobs(n_samples=500, n_features=2, centers=5, random_state=3)
```

```python
# fit the data into K-Means model
km.fit(X,y)
```

```python
# gives cluster center coordinates
centers = km.cluster_centers_
```

```python
# labels of each points
labels = km.labels_
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
plt.scatter(centers[:,0], centers[:,1], c='red', marker='^')
```
